```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd 
```

```python
#scrape a single game shots - input match id
base_url = 'https://understat.com/match/'
match = str(input('Please enter the match id: '))
url = base_url + match
```

```python
#Use requests to get the webpage & BeautifulSoup to parse the page
res = requests.get(url)
soup = BeautifulSoup(res.content,features='html')
scripts = soup.find_all('script')
```

```python
#get only the shotsData
strings = scripts[1].string
```

```python
strings
```

```python
#strip unnecessary symbols and get only JSON data
ind_start = strings.index("('")+2
ind_end = strings.index("')")
json_data = strings[ind_start:ind_end]
json_data = json_data.encode('utf8').decode('unicode_escape')

#convert string into JSON format
data = json.loads(json_data)
```

```python
data
```

```python
team = []
player = []
result = []
shotType = []
situation = []
minute = []
player_assisted = []
x = []
y = []
xG = []
result = []

data_away = data['a']
data_home = data['h']

for index in range(len(data_home)):
    for key in data_home[index]:
        if key == 'X':
            x.append(data_home[index][key])
        if key == 'Y':
            y.append(data_home[index][key])
        if key == 'h_team':
            team.append(data_home[index][key])
        if key == 'xG':
            xG.append(data_home[index][key])
        if key == 'result':
            result.append(data_home[index][key])
        if key == 'player':
            player.append(data_home[index][key])
        if key == 'shotType':
            shotType.append(data_home[index][key])
        if key == 'situation':
            situation.append(data_home[index][key])
        if key == 'minute':
            minute.append(data_home[index][key])
        if key == 'player_assisted':
            player_assisted.append(data_home[index][key])                                                
for index in range(len(data_away)):
    for key in data_away[index]:
        if key == 'X':
            x.append(data_away[index][key])
        if key == 'Y':
            y.append(data_away[index][key])
        if key == 'a_team':
            team.append(data_away[index][key])
        if key == 'xG':
            xG.append(data_away[index][key])
        if key == 'result':
            result.append(data_away[index][key])
        if key == 'player':
            player.append(data_away[index][key])
        if key == 'shotType':
            shotType.append(data_away[index][key])
        if key == 'situation':
            situation.append(data_away[index][key])
        if key == 'minute':
            minute.append(data_away[index][key])
        if key == 'player_assisted':
            player_assisted.append(data_away[index][key])    
```

```python
col_names = ['team','player','shotType','situation','minute','player_assisted','x','y','xG','result']
df = pd.DataFrame([team,player,shotType,situation,minute,player_assisted,x,y,xG,result],index=col_names)
df = df.T
```
